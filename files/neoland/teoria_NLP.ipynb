# Stemming

Reducir palabras a su raíz (puede no tener significado):
 + Organise, organising, organisation --> organis
 + intelligence, intelligently --> intelligen

Para hacer stemming usamos el algoritmo *Porter stemmer*
https://tartarus.org/martin/PorterStemmer/

````python
from nltk.stem import PorterStemmer
````

````python
texto
````

````python
frases = sent_tokenize(texto)
stemmer = PorterStemmer()
nueva_frase = []

for i in range(len(frases)):
    words = word_tokenize(frases[i])
    words=[stemmer.stem(word) for word in words]
    nueva_frase.append(' '.join(words))
    
' '.join(nueva_frase)
````

## Lematización
Similar al stemming, pero ahora la raíz sí que tiene significado:
+ going, goes, gone --> go
+ intelligence, intelligently --> intelligent

````python
nltk.download('wordnet')
````

````python
from nltk.stem import WordNetLemmatizer
````

https://es.wikipedia.org/wiki/WordNet

````python
frases = sent_tokenize(texto)
lemmatizer = WordNetLemmatizer()
nueva_frase = []

for i in range(len(frases)):
    words = word_tokenize(frases[i])
    words=[lemmatizer.lemmatize(word) for word in words]
    nueva_frase.append(' '.join(words))
    
nueva_frase
````

````python
texto2 = 'John is an intelligent individual. He intelligently does smart work. He is a top performer in the company'
texto2=' '.join(x.lower() for x in texto2.split())
````

## Análisis de sentimiento

El proceso de identificar y clasificar automáticamente opiniones expresadas en un texto, especialmente para determinar la actitud del autor hacia cierto tema.

Vamos a importar la base de datos `Sentiment.csv` de https://www.kaggle.com/crowdflower/first-gop-debate-twitter-sentiment .

````python
import pandas as pd
datos=pd.read_csv('Sentiment.csv')
````

````python
datos.head()
````

Seleccionamos sólo las columnas que nos interesan, `text` (será como nuestras 'features') y `sentiment` (será nuestro 'target').

````python
datos = datos[['text','sentiment']]
datos.head()
````

### Preprocesamiento

````python
datos.iloc[0,0]
````

Usando regex puedo limpiar los tweets de signos de puntuación, @, RTs y demás...

https://docs.python.org/3/library/re.html

https://stackoverflow.com/questions/27195769/python-remove-word-start-with-specific-character

````python
# Eliminamos los hashtags
datos['text']=datos['text'].str.replace('(\s)#\w+','')
datos.head()
````

````python
# Eliminamos las menciones
datos['text']=datos['text'].str.replace('(\s)@\w+','')
datos.head()
````

````python
# Quitamos los RT
datos['text']=datos['text'].str.replace('RT:','')
datos.head()
````

````python
# Dejo el texto en minúsculas
datos['text']=datos['text'].apply(lambda x: ' '.join(p.lower() for p in x.split()))
datos.head()
````

````python
# Quito las stopwords
datos['text']=datos['text'].apply(lambda x: ' '.join(p for p in x.split() if p not in stop_words))
datos.head()
````

````python
# Elimino la puntuación
datos['text']=datos['text'].str.replace('[^\w\s]','')
datos.head()
````

### Modelización

Por sencillez de la explicación, voy a considerar sólo tweets positivos y negativos:

````python
data = datos[datos['sentiment'] != 'Neutral']
data.head()
````

````python
data['sentiment'].value_counts()
````

Voy a separar en train y test, guardándome el 10% para test.

````python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.1)
````

Voy a obtener la tf-idf de cada término. (https://es.wikipedia.org/wiki/Tf-idf )

````python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1500, lowercase=True, analyzer='word', stop_words='english', ngram_range=(1,1))

train_tfidf = tfidf.fit_transform(train['text'])
````

````python
print(train_tfidf)
````

````python
train_tfidf
````

Nos da una `sparse matrix`.

Es una matriz normal, sus filas corresponden a los tweets y sus columnas a cada una de las posibles palabras, por ejemplo:

````python
tfidf.get_feature_names()[633]
````

El término $(d,t)$ es $\mathrm{tfidf}(t,d)$. Como casi todo son ceros en esta matriz, es más sencillo decir simplemente lo que valen los términos que no son ceros, esto es una `sparse matrix`. Esta matriz será el '$X$' de nuestro modelo.

Nuestro modelo será un Clasificador Bayesiano Ingenuo:

````python
from sklearn.naive_bayes import GaussianNB

# Definimos
gnb=GaussianNB()
# Ajustamos
gnb.fit(train_tfidf.toarray(),train['sentiment'])
````

Para predecir, hay que sacar también la tf-idf del test set:

````python
tfidf = TfidfVectorizer(max_features=1500, lowercase=True, analyzer='word', stop_words='english', ngram_range=(1,1))

test_tfidf = tfidf.fit_transform(test['text'])
````

````python
# Predecimos
pred = gnb.predict(test_tfidf.toarray())
````

````python
test['sentiment_pred']=pred
test
````

````python
test['sentiment'].value_counts()
````

````python
print('Número de puntos mal asignados de un total de %d puntos: %d' % (len(pred),(pred!= test['sentiment']).sum()) )
````
